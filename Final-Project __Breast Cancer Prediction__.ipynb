# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings("ignore")
plt.rcParams["figure.figsize"]=[10,5]
np.set_printoptions(suppress=True)
pd.options.display.float_format ='{:.2f}'.format
```

```python
original_data = pd.read_csv('Breast Cancer.csv')
original_data.head()
```

```python
cancer = original_data.copy(deep = True)
cancer.head()
```

# Data Preprocessing
#Cheking for Null values

```python
cancer.isnull().sum()
```

```python
print(cancer.columns)
print("\n\n")
print(f'Total Number of Columns : {len(cancer.columns)}')
```

```python
# Droping Unnamed : 32 column hence it won't consists of any data poits
cancer.drop(['Unnamed: 32'],axis = 1,inplace = True)
print(f'Total Number of Columns : {len(cancer.columns)}')
```

```python
cancer.isnull().sum().sum()
```

```python
sns.countplot(cancer['diagnosis'])
```

```python
#Here I am replacing the values M(Malignant) to 1 and B(Benign) to 0
cancer1 = cancer.copy(deep=1)
cancer1['diagnosis'].replace({'M':1,'B':0},inplace = True)
cancer1['diagnosis'].unique()

cancer1['diagnosis'].value_counts()
```

```python
cancer1['diagnosis'].info()
```

```python
cancer1.describe()
```

```python
# Here we are using seaborn pairplot to plot pairwise relationships between variables within a dataset
sns.pairplot(cancer1.iloc[:,0:5],hue ='diagnosis')
```

```python
# Here we can check the correlation of the dataset how they are correated with two or more continues variables
plt.figure(figsize= (25,10))
sns.heatmap(cancer1.corr(),annot = True,cmap= "cubehelix_r")
```

```python
# Here the machine learning model building comes in we need to split our dataset into train and test dato to build the model 
# Importing the Necessary Libraries for Model prediction 
import sklearn
from sklearn.model_selection import train_test_split

X = cancer1.drop(['id','diagnosis'],axis = 1)

Y = cancer1['diagnosis']

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state = 2)
```

```python
X.head()
```

```python
Y.head()
```

```python
#Here we are doing feature scaling to normalize the values in the independent variables 
# importing the library to do feature scaling 
from sklearn.preprocessing import StandardScaler
feature_scaling = StandardScaler()
x_train= feature_scaling.fit_transform(x_train)
x_test = feature_scaling.fit_transform(x_test)
x_train
```

```python
x_test
```

```python
#Model testing with SVM # classification_Report 
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
SVM = SVC()
SVM.fit(x_train,y_train)
predicted_xtest = SVM.predict(x_test)
predicted_xtrain = SVM.predict(x_train)
print(classification_report(y_test,predicted_xtest))
```

# Training Data Accuracy 

```python
print(accuracy_score(y_train,predicted_xtrain))
```

# Testing Data Accuracy

```python
# Accuracy
print(accuracy_score(y_test,predicted_xtest))
```

Obervation : -  Here we get 98% of accuracy with the SVM model and both the train and test data was perfectly fitted hence will try to improve the accuracy by ensemble techniques

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier


test_acc=0
train_acc=0
for i in range(10,100):
    RTC = RandomForestClassifier(n_estimators = 15,random_state = 2)
    RTC.fit(x_train,y_train)
    predicted_xtest = RTC.predict(x_test)
    predicted_xtrain = RTC.predict(x_train)
    if test_acc > accuracy_score(y_test,predicted_xtest):
        break
    else:
        test_acc = accuracy_score(y_test,predicted_xtest)

print(f"RandomForestClassifier Accuracy for training the data   {accuracy_score(y_train,predicted_xtrain)}")
print("\n\n")
print(f"RandomForestClassifier Accuracy for  testing the data   {accuracy_score(y_test,predicted_xtest)}")
```

# Bagging Technique

```python
from sklearn.ensemble import BaggingClassifier
BC = BaggingClassifier(DecisionTreeClassifier(),max_samples=0.5,max_features=0.6,n_estimators=15,random_state = 2)
BC.fit(x_train,y_train)
predicted_xtest = BC.predict(x_test)
predicted_xtrain = BC.predict(x_train)
print(f"Bagging Accuracy for training the data   {accuracy_score(y_train,predicted_xtrain)}")
print("\n\n")
print(f"Bagging Accuracy for testing the data   {accuracy_score(y_test,predicted_xtest)}")
```

# Boosting Technique

```python
from sklearn.ensemble import AdaBoostClassifier
ABC = AdaBoostClassifier(DecisionTreeClassifier(),n_estimators=20,random_state = 2)
ABC.fit(x_train,y_train)
predicted_xtest = ABC.predict(x_test)
predicted_xtrain = ABC.predict(x_train)
print(f"Boosting Accuracy for training the data   {accuracy_score(y_train,predicted_xtrain)}")
print("\n\n")
print(f"Boosting Accuracy for testing the data   {accuracy_score(y_test,predicted_xtest)}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GBC = GradientBoostingClassifier(n_estimators = 100,random_state = 2 )
GBC.fit(x_train,y_train)
predicted_xtest = GBC.predict(x_test)
predicted_xtrain = GBC.predict(x_train)
print(f"GradientBoostingClassifier Accuracy for training the data   {accuracy_score(y_train,predicted_xtrain)}")
print("\n\n")
print(f"GradientBoostingClassifier Accuracy for testing the data   {accuracy_score(y_test,predicted_xtest)}")
```
